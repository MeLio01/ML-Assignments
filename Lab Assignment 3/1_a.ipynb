```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Generate data
x,y=make_blobs(n_samples=100,n_features=2,centers=2,random_state=20)
x=pd.DataFrame(x)
x[2]=1
```

```python
sns.scatterplot(x=x.iloc[:,0],y=x.iloc[:,1],hue=y)
```

```python
def perceptron(x,y):
    m=len(x)
    n=len(x.columns)
    j=m
    k=100
    w=np.random.rand(n)
    #Loop until convergence or max iterations k
    while(k!=0 and j!=0):
        k-=1
        j=0
        plt.scatter(x.iloc[:,0],x.iloc[:,1],c=y,cmap='coolwarm')  
        plt.plot(x.iloc[:,0],-(w[0]*x.iloc[:,0]+w[2])/w[1],'r')
        plt.show()
        for i in range(m):
            dw=0
            if((w.T@x.iloc[i,:])<0 and y[i]==1):
                dw=x.iloc[i,:]
                j+=1
            elif((w.T@x.iloc[i,:])>=0 and y[i]==0):
                dw=-x.iloc[i,:]
                j+=1
            w=w+dw
    return w
```

```python
def h(z):
    if z>=0:
        return 1
    else:
        return 0
```

```python
w=perceptron(x,y)

```
