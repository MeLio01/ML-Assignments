```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
a= 10*np.random.rand()
b= 10*np.random.rand()
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
x = np.random.uniform(0,10,500)
y =  a*x + b + normal_dist(x,0,1)
plt.plot(x,y,'o')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
## DEFINE H(X) FOR LINEAR REGRESSION
def h(x,theta):
    return x@theta.T
```

```python
def cost_function(x,y,theta):
    m = len(x)
    J = (1/(2*m))*np.sum((h(x,theta)-y)**2)
    return J
    
```

```python
## Define gradient descent for linear regression
def gradient_descent(x,y,theta,learn_rate,epochs):
    m = len(x)
    J_all=[]
    epochs1=[]
    for i in range(epochs):
        z= (x.T)@(h(x,theta)-y)
        z= z.T
        theta = theta - (learn_rate/m)*z
        J_all.append(cost_function(x,y,theta))
        epochs1.append(i)
    plt.plot(epochs1,J_all)
    plt.xlabel("Epochs")
    plt.ylabel("Cost")
    return theta
```

```python
## Creating dataset
def create_points(degree,x):
    points = []
    for i in range(degree+1):
        points.append(x**i)
    return points
```

```python
degree =2
X_data = create_points(degree,x)
X_data = np.array(X_data).T
Y_data = y
theta = 10*abs((np.random.randn(degree+1)))
```

```python
## Normalizing X_data and Y_data
X_data = (X_data-np.mean(X_data))/np.std(X_data)
Y_data = (Y_data - np.mean(Y_data))/np.std(Y_data)
```

```python
train_x, test_x, train_y, test_y = train_test_split(X_data, Y_data, test_size=0.2)
theta_new = gradient_descent(train_x,train_y,theta,0.01,1000)
total_loss = cost_function(test_x,test_y,theta_new)
print(total_loss)
```
